# Classificando as variáveis qualitativas

```python
dados01.SEXO = dados01.SEXO.astype('category')
dados01.COR = dados01.COR.astype('category')
dados01.ALFABETIZADO = dados01.ALFABETIZADO.astype('category')
dados01.ESCOLA_CRECHE = dados01.ESCOLA_CRECHE.astype('category')
dados01.PEA = dados01.PEA.astype('category')

dados01.SEXO.cat.categories = ('Masculino','Feminino')
dados01.COR.cat.categories = ('Indígena','Branca','Preta','Amarela','Parda')
dados01.PEA.cat.categories = ('Economicamente ativas','Não economicamente ativas','Não se aplica')
dados01.ALFABETIZADO.cat.categories = ('Ignorado','Sim','Não')
```

```python
dados14.SEXO = dados14.SEXO.astype('category')
dados14.COR = dados14.COR.astype('category')
dados14.ALFABETIZADO = dados14.ALFABETIZADO.astype('category')
dados14.ESCOLA_CRECHE = dados14.ESCOLA_CRECHE.astype('category')
dados14.PEA = dados14.PEA.astype('category')

dados14.SEXO.cat.categories = ('Masculino','Feminino')
dados14.COR.cat.categories = ('Indígena','Branca','Preta','Amarela','Parda')
dados14.PEA.cat.categories = ('Economicamente ativas','Não economicamente ativas')
dados14.ALFABETIZADO.cat.categories = ('Sim','Não')
```

# Regiões 

### Analisando as frequencias relativas

```python
print('2001 - Frequências relativas por Região:')
ut1_01 = (dados01.REGIAO.value_counts(True) * 100)
print(ut1_01,'\n')

print('2014 - Frequências relativas por Região:')
ut1_14 = (dados14.REGIAO.value_counts(True) * 100)
print(ut1_14,'\n')
```

# Idade vs. Rendimento Mensal Total para diferentes anos de estudo


## 2001

```python
anos01_1 = dados01[(500000 > dados01.REND_MEN_TOT) & (dados01.REND_MEN_TOT > 0) & (dados01.IDADE < 999) & (dados01.ANOS_ESTUDO == 15)].IDADE
valor01_1 = dados01[(500000 > dados01.REND_MEN_TOT) & (dados01.REND_MEN_TOT > 0) & (dados01.IDADE < 999) & (dados01.ANOS_ESTUDO == 15)].REND_MEN_TOT

anos01_2 = dados01[(500000 > dados01.REND_MEN_TOT) & (dados01.REND_MEN_TOT > 0) & (dados01.IDADE < 999) & (dados01.ANOS_ESTUDO == 1)].IDADE
valor01_2 = dados01[(500000 > dados01.REND_MEN_TOT) & (dados01.REND_MEN_TOT > 0) & (dados01.IDADE < 999) & (dados01.ANOS_ESTUDO == 1)].REND_MEN_TOT


#Plotando
graf_1 = plt.plot(anos01_2,np.log(valor01_2), 'ro')
plt.title('Idade e Rendimento Mensal Total - 2001 \n Sem Instrução ou <1 Ano de estudo')
plt.xlabel('Idade')
plt.ylabel('log(Rendimento Mensal Total)')
plt.show()

graf_2 = plt.plot(anos01_1, np.log(valor01_1), 'bo')
plt.title('Idade e Rendimento Mensal Total - 2001 \n 14 anos de estudos')
plt.xlabel('Idade')
plt.ylabel('log(Rendimento Mensal Total)')
plt.show()
```

```python
print("Covariância entre Idade e Rendimento Mensal Total - 2001 - 14 anos de estudos :",np.cov(anos01_1,valor01_1)[0,1])
print("Correlação entre Idade e Rendimento Mensal Total - 2001 - 14 anos de estudos:",np.corrcoef(anos01_1,valor01_1)[0,1])
print()
print("Covariância entre Idade e Rendimento Mensal Total - 2001 - Menos que 1 ano de estudo:",np.cov(anos01_2,valor01_2)[0,1])
print("Correlação entre Idade e Rendimento Mensal Total - 2001 - Menos que 1 ano de estudo:",np.corrcoef(anos01_2,valor01_2)[0,1])
```

# Conclusão


Analisando os graficos e índices obtidos, podemos afirmar que:
    - A concrentração dos aposentados entre as regiões permaneceu praticamente igual.
    - A divisão de sexos sofreu uma inversão, em 2014 a maioria passou a ser aposentadas.
    - A divisão de raças/cor dos aposentados permaneceu-se praticamente inalterada, houve apenas um aumento da quantidade de pardos e diminuição da quantidade de aposentados
    - Os valores de aposentadoria, apesar de ter uma média maior em 2014, em relação ao salário mínimo de cada época apresentam uma distribuição bastante semelhante, com uma concentração maior entre 1 e 3 salários.
    - A distribuição de idade dos aposentados é bastante semelhante nos dois anos
    -A relação entre Anos de Estudo e Rendimento Mensal Total parece ter perdido força em 2014, visto a distribuição dos pontos dos gráficos e a variação dos índices de Correlação e Covariância.
